```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
digits
```

```python
digits.data.shape
```

```python
X=digits.data
#size 64 8*8
X
```

```python
digits.target.shape
y=digits.target
y
```

```python
#lets display
plt.figure(figsize=(20,4))
for index,(image,label) in enumerate(zip(digits.data[0:20],digits.target[0:20])):
    plt.subplot(1,20, index+1)
    plt.imshow(np.reshape(image,(8,8)),cmap=plt.cm.gray)
    plt.title(label,size=15)
    #our training data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=0)
```

```python
print("Train input data",X_train.shape)
print("Test input data",X_test.shape)
print("Train input data",y_train.shape)
print("Test input data",y_test.shape)

```

```python
# now it became 2d array
```

```python

log_reg=LogisticRegression()
#fit the model
log_reg.fit(X_train,y_train)
log_reg
```

```python
log_reg.predict(X_test[0:10])
predictions=log_reg.predict(X_test)
predictions
```

```python
#accuracy prediction/test/checking
score=log_reg.score(X_test,y_test)
print("the accuracy score is ",score)
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import confusion_matrix
cm=metrics.confusion_matrix(y_test, predictions)
cm
```

```python
#graph
import seaborn as sns
plt.figure(figsize=(9,9))
sns.heatmap(cm,annot=True,fmt='0.3f',linewidths=.5, square=True,cmap='Spectral');
plt.ylabel('Actual output',size=12)
plt.xlabel('Predicted output',size=12)
all_sample_title='Accuracy score:{0}'.format(score)
plt.title(all_sample_title, size=15)
```
